```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.Message,df.Category,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
```

```python
emails = [
            'Upto 20% discount reward!',
            'Hey mohan, discount Upto'
         ]
emails_cv=v.fit_transform(emails)
emails_cv
```

```python
v=CountVectorizer()
train_cv=v.fit_transform(X_train)
train_cv
```

```python
train_cv.toarray()
```

```python
v.get_feature_names_out()
```

```python
train_cv.shape
```

```python
pd.DataFrame(train_cv.toarray(),columns=v.get_feature_names_out())
```

```python
test_cv=v.transform(X_test)
test_cv
```

```python
test_cv.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(train_cv, y_train)
```

```python
model.score(test_cv, y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(test_cv)

print(classification_report(y_test, y_pred))
```
